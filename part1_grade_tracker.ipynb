```python
##task-1
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:

    name = student["name"].strip().title()
    roll = int(student["roll"])
    marks = [int(m) for m in student["marks_str"].split(", ")]

    cleaned = {"name": name, "roll": roll, "marks": marks}
    cleaned_students.append(cleaned)


    is_valid = all(word.isalpha() for word in name.split())
    validity = "✓ Valid name" if is_valid else "✗ Invalid name"


    print("================================")
    print(f"Student : {name}   {validity}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("================================")
    print()


for student in cleaned_students:
    if student["roll"] == 103:
        print(f"ALL CAPS  : {student['name'].upper()}")
        print(f"lowercase : {student['name'].lower()}")

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

##task-02

def get_grade(mark):
    if mark >= 90:
        return "A+"
    elif mark >= 80:
        return "A"
    elif mark >= 70:
        return "B"
    elif mark >= 60:
        return "C"
    else:
        return "F"

print(f"Student: {student_name}\n")
print(f"{'Subject':<12} {'Marks':>5}  {'Grade':>5}")
print("-" * 28)
for subject, mark in zip(subjects, marks):
    print(f"{subject:<12} {mark:>5}  {get_grade(mark):>5}")


total   = sum(marks)
average = round(total / len(marks), 2)

highest_idx = marks.index(max(marks))
lowest_idx  = marks.index(min(marks))

print(f"\nTotal Marks   : {total}")
print(f"Average Marks : {average}")
print(f"Highest       : {subjects[highest_idx]} ({marks[highest_idx]})")
print(f"Lowest        : {subjects[lowest_idx]} ({marks[lowest_idx]})")


print("\n── Add New Subjects (type 'done' to stop) ──")

new_count = 0

while True:
    subject_input = input("Enter subject name: ").strip()

    if subject_input.lower() == "done":
        break

    if not subject_input:
        print("⚠ Subject name cannot be empty. Try again.")
        continue

    mark_input = input(f"Enter marks for {subject_input} (0–100): ").strip()


    if not mark_input.replace(".", "").isdigit():
        print("⚠ Invalid input — marks must be a number. Entry skipped.\n")
        continue

    new_mark = float(mark_input)


    if not (0 <= new_mark <= 100):
        print("⚠ Invalid input — marks must be between 0 and 100. Entry skipped.\n")
        continue

    subjects.append(subject_input.title())
    marks.append(int(new_mark))
    new_count += 1
    print(f"✓ {subject_input.title()} added successfully.\n")

updated_average = round(sum(marks) / len(marks), 2)

print(f"\nNew subjects added : {new_count}")
print(f"Updated average    : {updated_average}")
print(f"All subjects now   : {subjects}")

##task-03

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]


results = []

print(f"{'Name':<18}| {'Average':^7} | {'Status'}")
print("-" * 40)

for name, marks in class_data:
    average = round(sum(marks) / len(marks), 2)
    status  = "Pass" if average >= 60 else "Fail"
    results.append((name, average, status))
    print(f"{name:<18}| {average:^7.2f} | {status}")



passed = [r for r in results if r[2] == "Pass"]
failed = [r for r in results if r[2] == "Fail"]

topper        = max(results, key=lambda r: r[1])
class_average = round(sum(r[1] for r in results) / len(results), 2)

print(f"\nStudents Passed : {len(passed)}")
print(f"Students Failed : {len(failed)}")
print(f"Class Topper    : {topper[0]} ({topper[1]})")
print(f"Class Average   : {class_average}")

##task-04

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "


clean_essay = essay.strip()
print(f"Step 1 - Stripped:\n{clean_essay}\n")


print(f"Step 2 - Title Case:\n{clean_essay.title()}\n")


python_count = clean_essay.count("python")
print(f"Step 3 - Count of 'python': {python_count}\n")


replaced = clean_essay.replace("python", "Python 🐍")
print(f"Step 4 - Replaced:\n{replaced}\n")


sentences = clean_essay.split(". ")
print(f"Step 5 - Sentences list:\n{sentences}\n")


print("Step 6 - Numbered sentences:")
for i, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")

```
